```python
import torch
from torchvision import transforms
import matplotlib.pyplot as plt
import pandas as pd
import resnet_analysis_utils as utils
```

```python
TEST_DATA_ROOT = "/scratch/bell/dutta26/mvtec/bottle/test/"
GROUND_TRUTH_ROOT = "/scratch/bell/dutta26/mvtec/bottle/ground_truth/"
MEMORY_BANK_PATH = "memory_bank.pt"  # Assumes it's in the same directory
IMAGE_SIZE = 256
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

print(f"Using device: {DEVICE}")
```

```python
model, features_dict, memory_bank = utils.load_resnet_essentials(MEMORY_BANK_PATH, DEVICE)

eval_transforms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=utils.IMAGENET_MEAN, std=utils.IMAGENET_STD),
])
```

```python
good_image_path = "/scratch/bell/dutta26/mvtec/bottle/test/good/000.png"
orig_tensor, anom_map, score = utils.get_resnet_anomaly_details(
    model, features_dict, memory_bank, good_image_path, eval_transforms, DEVICE
)
utils.plot_heatmap_and_overlay(orig_tensor, anom_map, score, title_prefix="Good Sample")
```

```python
bad_image_path = "/scratch/bell/dutta26/mvtec/bottle/test/broken_large/000.png"
orig_tensor, anom_map, score = utils.get_resnet_anomaly_details(
    model, features_dict, memory_bank, bad_image_path, eval_transforms, DEVICE
)
utils.plot_heatmap_and_overlay(orig_tensor, anom_map, score, "Defective Sample (broken_large)")
```

```python
bad_image_path = "/scratch/bell/dutta26/mvtec/bottle/test/contamination/000.png"
orig_tensor, anom_map, score = utils.get_resnet_anomaly_details(
    model, features_dict, memory_bank, bad_image_path, eval_transforms, DEVICE
)
utils.plot_heatmap_and_overlay(orig_tensor, anom_map, score, "Contamination")
```

```python
results_df = utils.calculate_scores_for_dataset(
    model, features_dict, memory_bank, TEST_DATA_ROOT, eval_transforms, DEVICE
)
```

```python
utils.plot_roc_curve(results_df)  #An area of 1 !!!
```
